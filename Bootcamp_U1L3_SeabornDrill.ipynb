```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataExams = pd.read_csv('/Users/jasonpaik9/Documents/Thinkful/Bootcamp/Unit 1/StudentsPerformance.csv')
```

```python
sns.set(style="darkgrid")
g = sns.factorplot(y="math score", data=dataExams,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Scores")
g.set_xlabels("Gender")
plt.title('Math Test Scores by Gender')
plt.show()
sns.set(style="darkgrid")
g = sns.factorplot(y="math score", data=dataExams,
                   size=6, kind="swarm", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Scores")
g.set_xlabels("Gender")
plt.title('Math Test Scores by Gender')
plt.show()
sns.set(style="darkgrid")
g = sns.factorplot(y="math score", data=dataExams,
                   size=6, kind="box", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Scores")
g.set_xlabels("Gender")
plt.title('Math Test Scores by Gender')
plt.show()
sns.set(style="darkgrid")
g = sns.factorplot(y="math score", data=dataExams,
                   size=6, kind="strip", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Scores")
g.set_xlabels("Gender")
plt.title('Math Test Scores by Gender')
plt.show()
```

- The bar plots itself show just a general visualization of where the exam scores are for all genders in the dataset. It shows the levels of a good holistic level but never go really into detail or give it any context.
- Swarm plots do a good job showing where many of the data lies, from a count perspective, and the range to where these scores are at. I think this is a good visualization to display where these scores are locationally in the range given. It also shows frequency of scores but on a pleasant-to-see graph.
- Box plots show, like the previous section, a good graph to see the ranges of scores with descriptive statistics embedded in the graphs. It doesn't show much context much it shows good data visually to see on a general level.
- The strip graph shows data like the swarm graph but shows it less in a parabolic shape but shows it more on a frequency level the way the graph is formed. The more congested the spots are together, the more frequent they are and the strip graph shows it on a great visual level.
